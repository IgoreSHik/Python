```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tensorflow.keras.layers import Conv2D
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.datasets import mnist
import cv2
```

```python
image_path = r"C:\Users\User\Desktop\Python\Palais_Royal_Paris.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
if image is None:
    print(f"ERROR {image_path}")
else:
    kernel_horizontal = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
    edges_horizontal = cv2.filter2D(image, -1, kernel_horizontal)

    kernel_vertical = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
    edges_vertical = cv2.filter2D(image, -1, kernel_vertical)

    blur_filter = np.ones((5, 5), np.float32) / 25
    blurred_image = cv2.filter2D(image, -1, blur_filter)

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 4, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')

    plt.subplot(1, 4, 2)
    plt.imshow(edges_horizontal, cmap='gray')
    plt.title('Horizontal Edges')

    plt.subplot(1, 4, 3)
    plt.imshow(edges_vertical, cmap='gray')
    plt.title('Vertical Edges')

    plt.subplot(1, 4, 4)
    plt.imshow(blurred_image, cmap='gray')
    plt.title('Blurred Image')

    plt.show()
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1)).astype('float32') / 255
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1)).astype('float32') / 255

model = Sequential([
    Conv2D(filters=8, kernel_size=(3, 3), input_shape=(28, 28, 1), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(units=10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()

model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test))

test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc}")
```
