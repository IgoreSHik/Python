```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(r'C:\Users\User\Downloads\archive\chess_games.csv')
```

```python
print("Liczba wpisów:")
print(data.count())
```

```python
print("Podstawowe statystyki:")
print(data.describe())
```

```python
print("Mediana:")
print(data.median())
```

```python
print("Odchylenie standardowe:")
print(data.std())
```

```python
data['turns'].plot(kind='hist')
plt.xlabel('turns')
plt.ylabel('Częstotliwość')
plt.title('Histogram turns')
plt.show()
```

```python
plt.scatter(data['turns'], data['opening_moves'])
plt.xlabel('turns')
plt.ylabel('opening_moves')
plt.title('Wykres rozrzutu turns vs opening_moves')
plt.show()
```

```python
grouped_data = data.groupby('victory_status')
print("Średnie wartości:")
print(grouped_data.mean())
print()

print("Kwantyle:")
print(grouped_data.quantile())
print()
```

```python
grouped_data = data.groupby('rated')
print("Średnie wartości:")
print(grouped_data.mean())
print()

print("Kwantyle:")
print(grouped_data.quantile())
print()
```

```python
value_counts = data['victory_status'].value_counts()

plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%')
plt.title('Koniec gry')
plt.show()
```

```python
value_counts = data['winner'].value_counts()

plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%')
plt.title('Zwycierzca')
plt.show()
```
