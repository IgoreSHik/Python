```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def calculate(self, wejscie):
        answer = np.dot(wejscie, self.weights) + self.bias
        return 1 if answer > 0 else 0
```

```python
def drawNet1(x1, x2, o1, o2, o12, y):
    G = nx.DiGraph()
    G.add_node("x1")
    G.add_node("x2")
    G.add_node("1: -0.5")
    G.add_node("2: -0.5")
    G.add_node("12: -0.5")
    G.add_node("y")

    G.add_edge("x1", "1: -0.5", weight = "1")
    G.add_edge("x2", "2: -0.5", weight = "1")
    G.add_edge("x1", "2: -0.5", weight = "-1")
    G.add_edge("x2", "1: -0.5", weight = "-1")
    G.add_edge("1: -0.5", "12: -0.5", weight = "1")
    G.add_edge("2: -0.5", "12: -0.5", weight = "1")
    G.add_edge("12: -0.5", "y")
    G.number_of_nodes()

    pos = {}
    pos["x1"] = (0,0.25)
    pos["x2"] = (0,0.75)
    pos["1: -0.5"] = (0.33, 0.25)
    pos["2: -0.5"] = (0.33, 0.75)
    pos["12: -0.5"] = (0.66, 0.5)
    pos["y"] = (1, 0.5)

    color_map = []
    if (x1 == 0):
        color_map.append('red')
    else:
        color_map.append('green')

    if (x2 == 0):
        color_map.append('red')
    else:
        color_map.append('green')

    if (o1 == 0):
        color_map.append('red')
    else:
        color_map.append('green')

    if (o2 == 0):
        color_map.append('red')
    else:
        color_map.append('green')

    if (o12 == 0):
        color_map.append('red')
    else:
        color_map.append('green')

    if (y == 0):
        color_map.append('red')
    else:
        color_map.append('green')

    plt.figure(figsize=(15,7))

    nx.draw_networkx_nodes(G, pos, node_color=color_map)
    nx.draw_networkx_edges(G, pos)
    nx.draw_networkx_edge_labels(G, pos)
    nx.draw_networkx_labels(G, pos)
    plt.show()
```
