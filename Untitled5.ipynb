```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def generate_data():
    data = np.random.rand(100, 4)
    return data

def plot_hyperplane_2D(w, theta, ax):
    x = np.linspace(0, 1, 100)
    y = (-w[0] * x - theta) / w[1]
    ax.plot(x, y, label=f'H(w, θ), θ={theta}')

def plot_hyperplane_3D(w, theta, ax):
    x = np.linspace(0, 1, 100)
    y = np.linspace(0, 1, 100)
    x, y = np.meshgrid(x, y)
    z = (-w[0] * x - w[1] * y - theta) / w[2]
    ax.plot_surface(x, y, z, alpha=0.5, label=f'H(w, θ), θ={theta}')
```

```python
def project_on_hyperplane(x, w, theta):
    projection = x - np.dot(x, w) / np.dot(w, w) * w
    return projection

def reflect_on_hyperplane(x, w, theta):
    reflection = x - 2 * np.dot(x, w) / np.dot(w, w) * w
    return reflection
```

```python
data = generate_data()

w_2D = np.array([1, -1])
theta_2D = 0.5
fig_2D, ax_2D = plt.subplots()
ax_2D.scatter(data[:, 0], data[:, 1], label='Dane')
plot_hyperplane_2D(w_2D, theta_2D, ax_2D)
ax_2D.legend()
plt.title('Hiperpłaszczyzna w dwóch wymiarach')
plt.show()
```

```python
w_3D = np.array([1, -1, 2])
theta_3D = 0.5
fig_3D = plt.figure()
ax_3D = fig_3D.add_subplot(111, projection='3d')
ax_3D.scatter(data[:, 0], data[:, 1], data[:, 2], label='Dane')
plot_hyperplane_3D(w_3D, theta_3D, ax_3D)
plt.title('Hiperpłaszczyzna w trzech wymiarach')
plt.show()
```

```python
vector_to_project = np.array([0.2, 0.4, 0.6])
projection_result = project_on_hyperplane(vector_to_project, w_3D, theta_3D)
print(f'Wejściowy wektor: {vector_to_project}')
print(f'Rzut na hiperpłaszczyznę: {projection_result}')

reflection_result = reflect_on_hyperplane(vector_to_project, w_3D, theta_3D)
print(f'Odbicie na hiperpłaszczyznę: {reflection_result}')
```

```python
def visualize_eigenvectors(data, eigenvectors):
    selected_eigenvectors = eigenvectors[:, :2]
    line_direction = selected_eigenvectors[:, 0]
    line_point = np.mean(data, axis=0)

    plt.scatter(data[:, 0], data[:, 1], label='Data')
    plt.quiver(line_point[0], line_point[1], line_direction[0], line_direction[1], scale=2, color='red', label='EigenVectors')
    plt.title('Visualization of Eigenvectors')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()
```

```python
eigenvectors = np.array([[1, 1], [-1, 1]])

visualize_eigenvectors(data, eigenvectors)
```

```python
def generate_data(mean, cov, num_samples):
    return np.random.multivariate_normal(mean, cov, num_samples)

def fisher_criterion(w, cov1, cov2, mean1, mean2):
    return np.dot(w.T, (mean1 - mean2)) / (0.5 * (np.dot(np.dot(w.T, cov1), w) + np.dot(np.dot(w.T, cov2), w)))

def fisher_classifier(data, w, threshold):
    scores = np.dot(data, w)
    predictions = (scores >= threshold).astype(int)
    return predictions
```

```python
np.random.seed(42)
mean1 = np.array([1, 1])
cov1 = np.array([[1, 0.5], [0.5, 1]])
data1 = generate_data(mean1, cov1, 100)

mean2 = np.array([5, 5])
cov2 = np.array([[1, -0.5], [-0.5, 2]])
data2 = generate_data(mean2, cov2, 100)

cov1_inv = np.linalg.inv(cov1)
cov2_inv = np.linalg.inv(cov2)
w = np.dot(cov1_inv + cov2_inv, (mean1 - mean2))

w /= np.linalg.norm(w)

scores1 = np.dot(data1, w)
scores2 = np.dot(data2, w)
threshold = (np.mean(scores1) + np.mean(scores2)) / 2

predictions1 = fisher_classifier(data1, w, threshold)
predictions2 = fisher_classifier(data2, w, threshold)

plt.scatter(data1[:, 0], data1[:, 1], label='Class 1')
plt.scatter(data2[:, 0], data2[:, 1], label='Class 2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Fisher Classifier')
plt.legend()

arrow_start = np.mean(np.vstack([data1, data2]), axis=0)
plt.arrow(arrow_start[0], arrow_start[1], w[0], w[1], head_width=0.2, head_length=0.2, fc='red', ec='red')

plt.show()
```
