```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Z = [0, 0]
X = [1, 2]
W = [5, 1]
P = 1
```

```python
plt.plot([0, X[0]], [0, X[1]])
#plt.plot([0, W[0]], [0, W[1]])
x = np.linspace(-100.,100.)
plt.plot(x,x*W[0]+P)
plt.plot(x,-x*W[1]+P)

plt.plot(x, x*W[0]-3, linestyle='dotted')
plt.plot(x, -x*W[1]+3, linestyle='dotted')

plt.xlim(-2, 7) 
plt.ylim(-2, 7)
plt.show()
```

```python
x = np.array([1, 3])

A, B, C = 1, -1, 1

x_proj = x - ((A * x[0] + B * x[1] + C) / (A**2 + B**2)) * np.array([A, B])

x_refl = -x_proj

plt.figure()
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, x_proj[0], x_proj[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, x_refl[0], x_refl[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.show()
```
