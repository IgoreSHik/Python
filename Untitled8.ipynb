```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score, roc_auc_score
from sklearn.datasets import make_classification, make_regression
```

```python
X, y = make_regression(n_samples=1000, n_features=20, noise=0.1, random_state=451)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=451)
output = []
```

```python
def dropout(X_train, y_train, X_test, y_test, param):
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dropout(param))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='linear'))

    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    output.append("Dropout   - param: "+str(param)+"   MSE: " + str(mse))
```

```python
dropout(X_train, y_train, X_test, y_test, 0.01)
dropout(X_train, y_train, X_test, y_test, 0.10)
dropout(X_train, y_train, X_test, y_test, 0.25)
dropout(X_train, y_train, X_test, y_test, 0.5)
dropout(X_train, y_train, X_test, y_test, 0.75)
```

```python
def l1(X_train, y_train, X_test, y_test, param):
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation='relu', kernel_regularizer=tf.keras.regularizers.l1(0.01)))
    model.add(Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l1(param)))
    model.add(Dense(1, activation='linear'))

    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    output.append("L1(Ridge) - param: "+str(param)+"   MSE: " + str(mse))
```

```python
l1(X_train, y_train, X_test, y_test, 0.01)
l1(X_train, y_train, X_test, y_test, 0.10)
l1(X_train, y_train, X_test, y_test, 0.25)
l1(X_train, y_train, X_test, y_test, 0.5)
l1(X_train, y_train, X_test, y_test, 0.75)
```

```python
def l2(X_train, y_train, X_test, y_test, param):
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)))
    model.add(Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(param)))
    model.add(Dense(1, activation='linear'))

    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    output.append("L2(Lasso) - param: "+str(param)+"   MSE: " + str(mse))
```

```python
l2(X_train, y_train, X_test, y_test, 0.01)
l2(X_train, y_train, X_test, y_test, 0.10)
l2(X_train, y_train, X_test, y_test, 0.25)
l2(X_train, y_train, X_test, y_test, 0.5)
l2(X_train, y_train, X_test, y_test, 0.75)
```
