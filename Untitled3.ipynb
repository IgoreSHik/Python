# Zadanie 1
Zaimplementować regułę uczenia Widrowa-Hoffa (LMS). Algorytm powinien działać dla dowolnego
wymiaru danych uczących oraz dowolnej liczby neuronów w warstwie.


```python
import numpy as np
import matplotlib.pyplot as plt

class LMS:
    def __init__(self, n):
        self.weights = np.random.rand(n)     
        self.bias = np.random.rand()
        
    def calculateArray(self, wejscie):
        wyjscie = np.random.rand(len(wejscie))
        for i, data in enumerate(wejscie):
            wyjscie[i] = self.calculate(data)
        return wyjscie

    def calculate(self, wejscie):
        answer = np.dot(wejscie, self.weights) + self.bias
        return answer

    def train(self, wejscie, answer, trainRate, n):
        for _ in range(n):
            op = [np.outer(x, x) for x in wejscie]
            R = np.mean(op, axis=0)
            h = np.mean(np.diff(wejscie))
            for i, data in enumerate(wejscie):
                suggestion = self.calculate(data)
                error = answer[i] - suggestion
                self.weights += error * trainRate * (np.dot(np.linalg.inv(R),data)-h)
                self.bias += error * trainRate
```

```python
class Perceptron:
    def __init__(self, n):
        self.weights = np.random.rand(n)
        self.bias = np.random.rand()
        
    def calculateArray(self, wejscie):
        wyjscie = np.random.rand(len(wejscie))
        for i, data in enumerate(wejscie):
            wyjscie[i] = self.calculate(data)
        return wyjscie

    def calculate(self, wejscie):
        answer = np.dot(wejscie, self.weights) + self.bias
        return answer

    def train(self, wejscie, answer, trainRate, n):
        print(self.bias)
        for _ in range(n):
            for i in range(len(wejscie)):
                suggestion = self.calculate(wejscie[i])
                error = answer[i] - suggestion
                self.weights += trainRate * error * wejscie[i]
                self.bias += trainRate * error
        print(self.bias)
```
